```python
import pygame
import sys
import random

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
WHITE = (255, 255, 255)
RED = (255, 0, 0)
# Set up the screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Wild Animal Shooting Game")

# Load gun images
gun_images = [
    pygame.image.load("rifle.jpg"),  # Replace with your gun image filenames
    pygame.image.load("rifle2.jpg"),
    # Add more images as needed...
]

# Rescale the gun images
scaled_gun_images = [pygame.transform.scale(img, (80, 40)) for img in gun_images]
#############3
forest_image = pygame.image.load("f.jpg") 
forest_image = pygame.transform.scale(forest_image, (WIDTH, HEIGHT))

#############
# Load animal images
animal_images = [
    pygame.image.load("lion55.png"),  # Replace with your animal image filenames
    pygame.image.load("ele12.jpg"),
    # Add more images as needed...
]

# Rescale the animal images
scaled_animal_images = [pygame.transform.scale(img, (50, 50)) for img in animal_images]

# Game variables
player_x, player_y = WIDTH // 2, HEIGHT - 100
bullet_speed = 5
bullets = []
animals = []
score = 0

clock = pygame.time.Clock()

# Function to spawn animals
def spawn_animal():
    animal_x = random.randint(50, WIDTH - 50)
    animal_y = random.randint(50, HEIGHT - 400)
    animal_image = random.choice(scaled_animal_images)
    return {'rect': pygame.Rect(animal_x, animal_y, 50, 50), 'image': animal_image}

# Game Loop
selected_gun = 0
running = True
while running:
    #screen.fill(WHITE)
    screen.blit(forest_image, (0, 0))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_1:
                selected_gun = 0
            elif event.key == pygame.K_2:
                selected_gun = 1
            # Add more keys for additional guns...

        if event.type == pygame.MOUSEBUTTONDOWN:
            bullet_x, bullet_y = player_x + 20, player_y
            bullets.append({'rect': pygame.Rect(bullet_x, bullet_y, 5, 15), 'speed': bullet_speed})

    mouse_x, mouse_y = pygame.mouse.get_pos()
    player_x = mouse_x - scaled_gun_images[selected_gun].get_width() / 2
    player_y = mouse_y - scaled_gun_images[selected_gun].get_height() / 2

    screen.blit(scaled_gun_images[selected_gun], (player_x, player_y))

    # Spawn animals
    if len(animals) < 5:  # Control the number of animals on screen
        animals.append(spawn_animal())

    # Move and draw animals
    for animal in animals:
        animal['rect'].y += 1  # Adjust the animal speed here
        screen.blit(animal['image'], (animal['rect'].x, animal['rect'].y))

    # Move and draw bullets
    for bullet in bullets:
        bullet['rect'].y -= bullet['speed']
        pygame.draw.rect(screen, (255, 0, 0), bullet['rect'])

    # Collision detection: Bullets hitting animals
    for animal in animals:
        for bullet in bullets:
            if animal['rect'].colliderect(bullet['rect']):
                animals.remove(animal)
                bullets.remove(bullet)
                score += 1

    # Display score
    font = pygame.font.Font(None, 36)
    score_text = font.render(f"Score: {score}", True, RED)
    screen.blit(score_text, (10, 10))
################
    if len(animals) == 0:
        game_over_text = font.render(f"Game Over! Total Score: {score}", True, RED)
        screen.blit(game_over_text, (WIDTH // 2 - 200, HEIGHT // 2))
#################
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```
