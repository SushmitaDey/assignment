```python
import pygame
import sys
import random
import math

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# Set up the screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Animal Shooting Game")


# Load the forest image
forest_image = pygame.image.load("f.jpg")  # Replace "forest.jpg" with your image file
# animal_images = [
#     pygame.image.load("animal33.jpg"),  # Replace these with your actual image filenames
#     pygame.image.load("animal44.jpg"),
#     # Add more images as needed...
# ]
# Rescale the forest image to fit the screen
forest_image = pygame.transform.scale(forest_image, (WIDTH, HEIGHT))
#scaled_animal_images = [pygame.transform.scale(img, (50, 50)) for img in animal_images]

# Fonts
font = pygame.font.Font(None, 36)

# Guns
guns = ["Pistol", "Shotgun", "Rifle"]
selected_gun = 0

# Player
player_radius = 15
player_x, player_y = WIDTH // 2, HEIGHT - 50
player_angle = 0
player_speed = 5

# Bullet
bullet_radius = 5
bullet_color = RED
bullets = []

# Animals
animal_width, animal_height = 50, 50
animal_speed = 3
animals = []
animal_spawn_timer = 100  # Timer to spawn animals
animal_spawn_counter = 0

score = 0

clock = pygame.time.Clock()

#Function to spawn animals
def spawn_animal():
    animal_x = random.randint(0, WIDTH - animal_width)
    animal_y = -animal_height
    return pygame.Rect(animal_x, animal_y, animal_width, animal_height)
    
# def spawn_animal():
#     animal_x = random.randint(0, WIDTH - 50)
#     animal_y = -50
#     animal_image = random.choice(scaled_animal_images)
#     return {'rect': pygame.Rect(animal_x, animal_y, 50, 50), 'image': animal_image}

# Function to draw the player
def draw_player(angle):
    pygame.draw.circle(screen, BLUE, (player_x, player_y), player_radius)
    end_x = player_x + int(math.cos(math.radians(angle)) * 2 * player_radius)
    end_y = player_y - int(math.sin(math.radians(angle)) * 2 * player_radius)
    pygame.draw.line(screen, BLUE, (player_x, player_y), (end_x, end_y), 2)

# Game Loop
running = True
animals=[]
gun_chosen = False
while running:
    #screen.fill(WHITE)
    screen.blit(forest_image, (0, 0))


    # # Move and draw animals
    # for animal in animals:
    #     animal['rect'].y += animal_speed
    #     screen.blit(animal['image'], (animal['rect'].x, animal['rect'].y))

    #     # Collision detection with bullets
    #     for bullet in bullets:
    #         if animal['rect'].colliderect(pygame.Rect(bullet[0], bullet[1], bullet_radius * 2, bullet_radius * 2)):
    #             # Handling bullet-animal collision...
    #             pass
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if not gun_chosen:
                if event.key == pygame.K_1:
                    selected_gun = 0
                    gun_chosen = True
                elif event.key == pygame.K_2:
                    selected_gun = 1
                    gun_chosen = True
                elif event.key == pygame.K_3:
                    selected_gun = 2
                    gun_chosen = True
            elif event.key == pygame.K_SPACE:
                bullet_angle = player_angle
                bullet_x = player_x
                bullet_y = player_y

                bullet_velocity = 10
                bullet_vel_x = math.cos(math.radians(bullet_angle)) * bullet_velocity
                bullet_vel_y = -math.sin(math.radians(bullet_angle)) * bullet_velocity

                bullets.append([bullet_x, bullet_y, bullet_vel_x, bullet_vel_y])

    if not gun_chosen:
        pistol_text = font.render("1 - Pistol", True, BLACK)
        shotgun_text = font.render("2 - Shotgun", True, BLACK)
        rifle_text = font.render("3 - Rifle", True, BLACK)
        screen.blit(pistol_text, (WIDTH // 2 - 50, HEIGHT // 2 - 50))
        screen.blit(shotgun_text, (WIDTH // 2 - 50, HEIGHT // 2))
        screen.blit(rifle_text, (WIDTH // 2 - 50, HEIGHT // 2 + 50))
    else:
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            player_angle += player_speed
        if keys[pygame.K_RIGHT]:
            player_angle -= player_speed

        # Spawn animals
        if animal_spawn_counter == animal_spawn_timer:
            animals.append(spawn_animal())
            animal_spawn_counter = 0
        else:
            animal_spawn_counter += 1

        # Move and draw animals
        for animal in animals:
            animal.y += animal_speed
            pygame.draw.rect(screen, GREEN, animal)


            # Collision detection with bullets
            for bullet in bullets:
                if animal.colliderect(pygame.Rect(bullet[0], bullet[1], bullet_radius * 2, bullet_radius * 2)):
                    animals.remove(animal)
                    bullets.remove(bullet)
                    score += 1

        # Remove bullets that go off-screen
        bullets = [[bullet[0], bullet[1], bullet[2], bullet[3]] for bullet in bullets if (0 <= bullet[0] <= WIDTH and 0 <= bullet[1] <= HEIGHT)]
        for bullet in bullets:
            bullet[0] += bullet[2]
            bullet[1] += bullet[3]
            
            pygame.draw.circle(screen, bullet_color, (int(bullet[0]), int(bullet[1])), bullet_radius)
        # Draw the player
        draw_player(player_angle)

        # Display score
        score_text = font.render(f"Score: {score}", True, BLACK)
        screen.blit(score_text, (10, 10))

        if len(animals) == 0:
            game_over_text = font.render(f"Game Over! Total Score: {score}", True, RED)
            screen.blit(game_over_text, (WIDTH // 2 - 200, HEIGHT // 2))

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```
